```
#box plot #violinplot histogram
sns.boxplot(data=df,x="status",y="age")
sns.set_style("dark")
```

```
sns.violinplot(data=df,x="status",y="age",set_style="whitegrid")
```

```
sns.FacetGrid(df,hue="status",size=5)\
.map(sns.distplot,"age")\
.add_legend
plt.show()
```

```
status_1=df[df["status"]==1]
```

```
status_2=df[df["status"]==2]
```

```
status_2
```

```
count,edges=np.histogram(status_1["age"],bins=10,density=True)
pdf=count/sum(count)
cdf=np.cumsum(pdf)
plt.plot(edges[1:],pdf,label="PDF for status1")
plt.plot(edges[1:],cdf,label="CDF for status1")
plt.show
#status_2
count1,bin_edges=np.histogram(status_2["age"],bins=5,density=True)
pdf=count1/(sum(count1))
pdf
cdf=np.cumsum(pdf)

plt.plot(bin_edges[1:],pdf,label="PDF for status2",color="r")
plt.plot(bin_edges[1:],cdf,label="CDF for status2",color="g")
plt.xlabel("CDF and PDF for age ")
plt.ylabel("Percentage")
plt.legend()
plt.show()
```

```
#PDF and CDF of feature nodes with respect to status

counts,edges=np.histogram(status_1["nodes"],bins=10)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(edges[1:],pdf,label="PDF for status1")
plt.plot(edges[1:],cdf,label="CDF for status1")

counts,edges=np.histogram(status_2["nodes"],bins=10)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(edges[1:],pdf,label="PDF for status2")
plt.plot(edges[1:],cdf,label="CDF for status2")
plt.legend()
plt.show()
```
